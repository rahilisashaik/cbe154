```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset-creating publishable-quality figures in Excel.xlsx - Sheet1.csv")
```

```python
df.head()
```

```python
x = [item * .001 for item in df["‚àÜp (Pa)"]]
y = [item * 1000 for item in df["J (mol / s m^2)"]]
print(x)
```

```python
coeffs = np.polyfit(x, y, 1)
fit_line = np.polyval(coeffs, x)

y_err = np.abs(y - fit_line)
x_err = np.abs(x - fit_line) * .001
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, fit_line, label = "Pure O2 - Co-current")
plt.errorbar(
    x, y, xerr=x_err, yerr=y_err, fmt='o', markersize="2", capsize=5, label='Data point errors',
    color='red', markerfacecolor='yellow', ecolor='blue'
)
plt.xlabel("‚àÜp (Pa) * 1000")
plt.ylabel("J (mol / s m^2) * .001")
plt.grid()
plt.legend()
```
